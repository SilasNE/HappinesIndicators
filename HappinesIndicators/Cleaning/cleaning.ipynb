```python
import pandas as pd

df = pd.read_csv("2015.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
relevante_spalten = [
    "Country",
    "Happiness Rank",
    "Happiness Score",
    "Standard Error",          # ← neu
    "Economy (GDP per Capita)",
    "Family",
    "Health (Life Expectancy)",
    "Freedom",
    "Generosity",
    "Trust (Government Corruption)",
    "Dystopia Residual"
]
df = df[relevante_spalten]
df.info()
```

```python
print(df.isnull().sum())          
```

```python
print(df.duplicated(subset="Country").sum())
```

```python
print(df.shape)

print(df["Happiness Score"].shape[0])
df = df[df["Happiness Score"] > 0]
print(df["Happiness Score"].shape[0])
print(df["Economy (GDP per Capita)"].shape[0])
df = df[df["Economy (GDP per Capita)"] >= 0]
print(df["Economy (GDP per Capita)"].shape[0])
print(df["Family"].shape[0])
df = df[df["Family"] >= 0]
print(df["Family"].shape[0])
print(df["Health (Life Expectancy)"].shape[0])
df = df[df["Health (Life Expectancy)"] >= 0]
print(df["Health (Life Expectancy)"].shape[0])
```

```python
numerische_spalten = [
    "Happiness Score", "Economy (GDP per Capita)", "Family",
    "Health (Life Expectancy)", "Freedom", "Generosity",
    "Trust (Government Corruption)"
]
df[numerische_spalten] = df[numerische_spalten].apply(pd.to_numeric, errors="coerce")
df = df.dropna() 
```

```python
df = df.reset_index(drop=True)

print(f"\nCleaning abgeschlossen: {len(df)} Länder, {len(df.columns)} Spalten")
```

```python
df.to_csv("2015_cleaned.csv", index=False)
```
